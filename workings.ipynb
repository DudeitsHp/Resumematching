```python
import pandas as pd
import re
print ("Import success")
```

```python
# Load your skills file
skills_df = pd.read_csv("skills.csv")
```

```python
# Clean and filter
skills_list = []
for skill in skills_df['skill']:
    if pd.isna(skill):
        continue
    skill = skill.strip().lower()
    
    # Skip short or non-informative words
    if len(skill) < 3:
        continue
    if skill in ["is", "be", "as", "do", "or", "and", "of", "to", "in", "for", "with", "by", "an", "on"]:
        continue
    if re.match(r"^[a-z]{1,2}$", skill):  # single/double letters like 'pf', 'll'
        continue
    if any(word in skill for word in ["degree", "bachelors", "communication", "good", "strong", "preferred", "knowledge", "fresher", "excellent", "internal", "relations", "methods"]):
        continue

    # Keep only meaningful entries
    skills_list.append(skill)
```
